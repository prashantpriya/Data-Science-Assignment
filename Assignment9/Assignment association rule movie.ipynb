```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset
df=pd.read_csv('my_movies.csv')
df
```

```python
df2=df.iloc[:,5:]
df2
```

```python
#Apriori Algorithm
# Association rules with 10% Support and 70% confidence
```

```python
# with 10% support
score = apriori(df2, min_support=0.1, use_colnames=True,verbose=1)
```

```python
score
```

```python
# 70% confidence
model=association_rules(score,metric='lift',min_threshold=0.7)
model
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
model[model.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(model['support'],model['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
#5% support and 80% confidence
```

```python
# with 5% support
score2=apriori(df2,min_support=0.05,use_colnames=True)
score2
```

```python
# 90% confidence
model2=association_rules(score2,metric='lift',min_threshold=0.8)
model2
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
model2[model2.lift>1]
```

```python
plt.scatter(model2['support'],model2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
